```python
import pandas as pd
import numpy as np
import scipy as sp

from itertools import chain

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##aa from string to different columns 
```

```python
#read in persona data
persona =pd.read_csv('~/Documents/persona_project_cardiology/original data and supplements/persona_exploration.csv')
```

```python
persona.shape
```

```python
#unique number of devices 
len(persona.DEVICE_ID_HASHED.drop_duplicates())
```

```python
#select AA column from persona data 
aa_persona=persona.AA
```

```python
#split AA by comma
aa_persona_a=[]
for i in range(len(aa_persona)):
    aa_persona_a.append(aa_persona[i].split(","))
```

```python
#unlist AA
aa_persona_b=list(chain(*aa_persona_a))

#change AA from list to array and drop duplicates
aa_persona_c=np.asarray(aa_persona_b)

aa_persona_d=np.unique(aa_persona_c)
```

```python
#put AA into a table 
aa_persona_df=pd.DataFrame(aa_persona_d)
aa_persona_df=aa_persona_df.rename({0: 'aa'}, axis='columns')
aa_persona_df.aa=aa_persona_df.aa.astype('int64')
```

```python
#create empty new dataframe with unique device as rows and unique aa as columns 
aa_1=aa_persona_df.aa.astype('str').drop_duplicates()
aa_str='ID_'+aa_persona_df.aa.astype('str').drop_duplicates()


aa_df=pd.DataFrame(0, index=persona.DEVICE_ID_HASHED, columns=aa_str)
```

```python
#fill in the AA table. 
for j in range(persona.shape[0]):
    for i in range(len(aa_1)):
        if aa_1.iloc[i] in persona.AA[j]:
            aa_df.iat[j,i]=1
        else:
            aa_df.iat[j,i]=0
            
#this table puts aa from its original long string format in persona data into a table with one aa one column
#it not only applies here
```

```python
aa_df.to_csv('~/Documents/persona_project_cardiology/clustering/AllUniqueAAToDevice.csv')
```
