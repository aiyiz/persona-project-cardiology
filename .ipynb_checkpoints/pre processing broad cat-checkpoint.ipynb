```python
#put these segment ID into one list
level3=list(chain(a,b,c,d,e,f,g,h,i,j,k,l,m,n))
```

```python
#take subset of dataframe for segments need the most granular info
l3_df=v3[v3['SEGMENT_ID'].isin(level3)]
```

```python
#take subset of dataframe for segments need the second to most granular info
l2_df=v3[v3['SEGMENT_ID'].isin(level3) == False]
```

```python
#create new variable called name for segment ID and name match incorporating different levels of granularity
l3_df['name']=l3_df['CATEGORY2']+" "+l3_df['CATEGORY3'].astype('str')+' '+l3_df['SEGMENT_NAME']
```

```python
#create new variable called name for segment ID and name match incorporating different levels of granularity
l2_df['name']=l2_df['CATEGORY2']+" "+l2_df['CATEGORY3'].astype('str')
```

```python
#put level3 and level2 together into one dataframe called v3b and create new segment ID called id_new base on new granularity info
v3b=l3_df.append(l2_df)
v3b['id_new'] = v3b.groupby(['name']).ngroup()
```

```python
broadCatName=v3b[['id_new','name']].drop_duplicates()
```
