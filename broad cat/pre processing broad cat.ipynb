```python
import pandas as pd
import numpy as np
import scipy as sp

from itertools import chain

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
aa_df=pd.read_csv('~/Documents/persona_project_cardiology/clustering/AllUniqueAAToDevice.csv')
```

```python
#read in v3 data which is a dictionary of aa and what do they stand for
v3b =pd.read_csv('~/Documents/persona_project_cardiology/clustering/v3b.csv')
```

```python
#dictionary
dict =pd.read_csv('~/Documents/persona_project_cardiology/clustering/broadCatNameDict.csv')
```

```python
#reshape aa df from wide to long 
aa_df2=pd.melt(aa_df, id_vars=['DEVICE_ID_HASHED'], value_vars=aa_df.columns.values[1:])
```

```python
aa_df2.columns=['DEVICE_ID_HASHED','SEG_ID_STR','Indicator']
```

```python
v3b['SEG_ID_STR']='ID_'+v3b['SEGMENT_ID'].astype(str)
```

```python
#merge with dictionary table to get device to multiple new id association
aa_df3=aa_df2.merge(v3b,left_on='SEG_ID_STR', right_on='SEG_ID_STR', how='left')
```

```python
aa_df4=aa_df3.groupby(['DEVICE_ID_HASHED','id_new'])[['Indicator']].sum()
```

```python
aa_df5 = pd.pivot_table(aa_df4, values='Indicator', index=['DEVICE_ID_HASHED'], columns=['id_new'], aggfunc=np.sum)
```

```python
aa_df5.head()
```

```python
def change(x): 
    if x>=1:
        x=1
    else:
        x=0
    return x
```

```python
aa_df6=aa_df5
```

```python
for i in range(aa_df6.shape[1]):
    aa_df6.iloc[:,i]=aa_df6.iloc[:,i].apply(change)
```
